```python
# ai_courses_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("ai_courses_uk_us.csv")

# Basic info
print("Dataset Info:")
print(df.info())
print("\nDataset Preview:")
print(df.head())

# Basic summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Sales Comparison by Country
plt.figure(figsize=(10,6))
sns.barplot(x="Country", y="Sales_USD_Millions", hue="Year", data=df)
plt.title("AI Course Sales by Country")
plt.ylabel("Sales (USD Millions)")
plt.show()

# Profit Margin distribution
plt.figure(figsize=(10,6))
sns.boxplot(x="Country", y="Profit_Margin_Percent", data=df)
plt.title("Profit Margin Distribution by Country")
plt.show()

# Demographic Age Distribution
plt.figure(figsize=(10,6))
sns.scatterplot(x="Avg_Learner_Age", y="Sales_USD_Millions", hue="Country", data=df)
plt.title("Correlation: Avg Learner Age vs Sales")
plt.show()

# Gender-Based Enrollment
gender_df = df[["Country","Percent_Female","Percent_Male"]].groupby("Country").mean().reset_index()
print("\nAverage Gender Enrollment by Country:")
print(gender_df)
```
