```python
# Requires: pandas, numpy
import pandas as pd
import numpy as np

# --------------- Configurable parameters -------------

# Countries of interest
countries = ["United Kingdom", "United States"]

# Years you want in your dataset
years = list(range(2019, 2024))   # example: last 5 years 2019-2023
# You can extend to 2024 or 2025 as needed

# Items/categories
items = ["Hash Browns", "Eggs"]

# Example demographic segments - adjust or expand as needed
demographic_segments = [
    "Age 18-34",
    "Age 35-54",
    "Age 55+",
    "Urban",
    "Suburban",
    "Rural"
]

# -------------------------------------------------------
# Helper function to simulate or structure data
def generate_placeholder_data():
    records = []
    rng = np.random.default_rng(seed=42)
    
    for country in countries:
        for year in years:
            for item in items:
                # Simulated sales units in thousands
                # Replace with real values later
                base_sales = 1000 if item == "Eggs" else 300
                sales_units = base_sales * (1 + 0.05 * (year - years[0]))  # small growth
                sales_units *= rng.normal(loc=1.0, scale=0.1)  # random noise
                
                # Simulated revenue in thousands USD or GBP
                # Choose currency or convert consistently outside this template
                price_per_unit = 0.15 if item == "Eggs" else 0.5
                revenue = sales_units * price_per_unit
                
                # Simulated profit margin in percent
                # Placeholder: eggs higher margin than hash browns in this simulation
                if item == "Eggs":
                    profit_margin = rng.normal(loc=12, scale=2)  # 12% avg
                else:
                    profit_margin = rng.normal(loc=8, scale=2)   # 8% avg
                
                # Demographic shares - must sum to <=1 across segments for each record
                # This example assigns random shares and then normalizes
                shares = rng.random(len(demographic_segments))
                shares /= shares.sum()
                
                # Build record
                record = {
                    "country": country,
                    "year": year,
                    "item": item,
                    "sales_units_k": round(sales_units, 1),
                    "revenue_k": round(revenue, 1),
                    "profit_margin_pct": round(profit_margin, 2),
                }
                
                # Add demographic shares
                for seg, share in zip(demographic_segments, shares):
                    record[f"share_{seg}"] = round(share, 4)
                
                records.append(record)
    
    df = pd.DataFrame.from_records(records)
    
    # Optional: reorder columns for readability
    cols = [
        "country", "year", "item",
        "sales_units_k", "revenue_k", "profit_margin_pct"
    ] + [f"share_{seg}" for seg in demographic_segments]
    df = df[cols]
    return df

# Generate placeholder dataset
df = generate_placeholder_data()

# Show a snippet
print(df.head(10))

# -------------------------------------------------------
# Save to CSV (optional)
# df.to_csv("hashbrowns_eggs_sales_demographics.csv", index=False)
```
