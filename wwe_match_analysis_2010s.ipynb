```python
# wwe_2016_analysis.py

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. LOAD DATA
# -----------------------------
def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print("✅ Data loaded successfully.\n")
        return df
    except Exception as e:
        print("❌ Error loading file:", e)
        return None


# -----------------------------
# 2. CLEAN DATA
# -----------------------------
def clean_data(df):

    print("🔍 Initial Shape:", df.shape)

    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

    # Remove duplicates
    df = df.drop_duplicates()

    # Handle missing values
    df = df.dropna(how='all')  # drop completely empty rows

    # Convert possible date column
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # Convert attendance to numeric if exists
    if 'attendance' in df.columns:
        df['attendance'] = pd.to_numeric(df['attendance'], errors='coerce')

    # Strip text columns
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.strip()

    print("✅ Data cleaned.")
    print("🔍 New Shape:", df.shape, "\n")

    return df


# -----------------------------
# 3. BASIC ANALYSIS
# -----------------------------
def basic_analysis(df):

    print("📊 BASIC STATISTICS\n")
    print(df.describe(include='all'))
    print("\n")

    # Matches by brand
    if 'brand' in df.columns:
        print("📌 Matches by Brand:")
        print(df['brand'].value_counts(), "\n")

    # Wins by wrestler
    if 'winner' in df.columns:
        print("🏆 Wins by Wrestler:")
        print(df['winner'].value_counts().head(10), "\n")

    # Attendance analysis
    if 'attendance' in df.columns:
        print("🎟 Attendance Summary:")
        print(df['attendance'].describe(), "\n")


# -----------------------------
# 4. VISUALIZATIONS
# -----------------------------
def visualize_data(df):

    # Brand distribution
    if 'brand' in df.columns:
        df['brand'].value_counts().plot(kind='bar')
        plt.title("Matches by Brand (2016)")
        plt.xlabel("Brand")
        plt.ylabel("Number of Matches")
        plt.tight_layout()
        plt.show()

    # Monthly matches
    if 'date' in df.columns:
        monthly_matches = df.groupby(df['date'].dt.month).size()
        monthly_matches.plot(kind='line', marker='o')
        plt.title("Matches Per Month (2016)")
        plt.xlabel("Month")
        plt.ylabel("Number of Matches")
        plt.tight_layout()
        plt.show()

    # Attendance distribution
    if 'attendance' in df.columns:
        df['attendance'].plot(kind='hist', bins=15)
        plt.title("Attendance Distribution")
        plt.xlabel("Attendance")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()


# -----------------------------
# 5. SAVE CLEANED DATA
# -----------------------------
def save_cleaned_data(df, output_file):
    df.to_csv(output_file, index=False)
    print(f"💾 Cleaned data saved as '{output_file}'\n")


# -----------------------------
# 6. MAIN FUNCTION
# -----------------------------
def main():

    file_path = "WWE-Data-2016.csv"
    output_file = "wwe_2016_cleaned.csv"

    df = load_data(file_path)

    if df is not None:
        df = clean_data(df)
        basic_analysis(df)
        visualize_data(df)
        save_cleaned_data(df, output_file)


if __name__ == "__main__":
    main()
```
