```python
import pandas as pd

# ---- Configuration: years, regions, family-size buckets ----
years = [2022, 2023, 2024]
regions = ["UK", "US"]

# family size buckets (example categories)
family_buckets = [
    "Single-person household",
    "Small family (2-3 persons)",
    "Medium family (4 persons)",
    "Large family (5+ persons)"
]

# ---- Synthetic baseline sales figures ----
# These are illustrative starting points; substitute your own data where available.
# For UK 2024 we anchor to public reporting that UK Easter egg spending was >£255m in 2024.
# For US 2024 we anchor to NCA reporting that Easter confectionery sales topped $5.39B in 2024.
# Note: US figure is broader than just eggs; adjust as desired for eggs-only share.

baseline_sales = {
    ("UK", 2024): 255_000_000,   # GBP, reported spending on Easter eggs in 2024 (public source)
    ("US", 2024): 5_390_000_000  # USD, reported Easter confectionery sales in 2024
}

# Assume modest growth or decline for other years; these are placeholders.
growth_factors = {
    ("UK", 2023): 1.05,  # 5% higher/lower relative to a base; adjust as needed
    ("UK", 2022): 1.03,
    ("US", 2023): 1.02,
    ("US", 2022): 1.00,
}

# Compute synthetic annual sales by region
sales_records = []
for region in regions:
    for year in years:
        key = (region, 2024)
        if (region, year) == (region, 2024):
            sales = baseline_sales[key]
        else:
            # back-calculate or project relative to 2024 figure
            # Here we simply apply a factor to make up sample values;
            # in a real use case, replace with actual historical data.
            factor = growth_factors.get((region, year), 1.0)
            sales = baseline_sales[key] * factor
        sales_records.append({
            "region": region,
            "year": year,
            "sales_value": sales
        })

sales_df = pd.DataFrame(sales_records)

# ---- Summary profit margins (illustrative) ----
# Summary-only estimates, e.g., typical gross margin range for confectionery or seasonal items.
# Replace or refine with your own or third‑party data where allowed.
margin_summary = {
    "UK": {
        "typical_gross_margin_pct": 30,   # example: 30%
        "note": "Summary estimate; actual varies by brand/retailer"
    },
    "US": {
        "typical_gross_margin_pct": 35,   # example: 35%
        "note": "Summary estimate; actual varies by brand/retailer"
    }
}

# ---- Family-size demographics shares (illustrative sample shares) ----
# The shares should total 100% per region/year; here assumed stable across years.
# Replace with actual survey or census inputs if available.
family_shares = {
    "UK": {
        "Single-person household": 30,
        "Small family (2-3 persons)": 35,
        "Medium family (4 persons)": 20,
        "Large family (5+ persons)": 15
    },
    "US": {
        "Single-person household": 28,
        "Small family (2-3 persons)": 40,
        "Medium family (4 persons)": 22,
        "Large family (5+ persons)": 10
    }
}

# ---- Build final dataset combining sales, margins, and family demographics ----
rows = []
for _, sales_row in sales_df.iterrows():
    region = sales_row["region"]
    year = sales_row["year"]
    sales_value = sales_row["sales_value"]
    margin = margin_summary[region]["typical_gross_margin_pct"]
    margin_note = margin_summary[region]["note"]
    for bucket in family_buckets:
        share_pct = family_shares[region][bucket]
        rows.append({
            "region": region,
            "year": year,
            "family_size_bucket": bucket,
            "family_share_pct": share_pct,        # share of households or consumers
            "sales_value": sales_value,           # total region sales (same for each bucket row)
            "estimated_margin_pct": margin,        # summary margin
            "margin_note": margin_note
        })

dataset_df = pd.DataFrame(rows)

# ---- Optional: compute weighted sales by family bucket ----
# Example: distribute sales proportionally by family share
dataset_df["sales_by_bucket"] = (
    dataset_df["sales_value"] * dataset_df["family_share_pct"] / 100.0
)

# ---- View sample ----
print(dataset_df.head(12))
```
