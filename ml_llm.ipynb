```python
# =============================================================================
# Jupyter Notebook: Analysis of Online ML & LLM Sales + Demographics
# =============================================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load dataset ---
df = pd.read_csv("online_ml_llm_sales_demographics.csv")

# --- Quick overview ---
print("Dataset shape:", df.shape)
print(df.head())

# --- Basic statistics ---
print("\nSummary statistics:")
print(df.describe(include='all'))

# --- Filter by Country ---
us_data = df[df["Country"] == "US"]
uk_data = df[df["Country"] == "UK"]

# --- Revenue Comparison ---
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="CourseType", y="SalesRevenue_USD", hue="Country")
plt.title("Sales Revenue by Course Type (UK vs US)")
plt.ylabel("Revenue (USD)")
plt.xlabel("Course Type")
plt.show()

# --- Profit Margin by Course Type ---
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="CourseType", y="ProfitMargin", hue="Country")
plt.title("Profit Margin by Course Type (UK vs US)")
plt.ylabel("Profit Margin")
plt.xlabel("Course Type")
plt.show()

# --- Learners by AgeGroup ---
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="AgeGroup", hue="Country")
plt.title("Learner Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Records")
plt.show()

# --- Gender Breakdown ---
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Gender", hue="Country")
plt.title("Learner Gender Distribution")
plt.show()

# --- Cross-Tab: Learners by EducationLevel & Country ---
edu_ct = pd.crosstab(df["EducationLevel"], df["Country"])
print("\nCross Tab of Education Level by Country:")
print(edu_ct)

# --- Show Profit vs Revenue scatter ---
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="SalesRevenue_USD",
    y="ProfitMargin",
    hue="Country",
    style="CourseType",
    s=100
)
plt.title("Profit Margin vs Sales Revenue")
plt.xlabel("Sales Revenue (USD)")
plt.ylabel("Profit Margin")
plt.show()
```
